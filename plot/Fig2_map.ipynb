```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import xarray as xr
import glob 

import os
os.environ['PROJ_LIB'] = '/home/Youtong.Zheng/.conda/envs/demo/share/proj'
from mpl_toolkits import basemap
from mpl_toolkits.basemap import Basemap,maskoceans 

from plot_utils import pltpolar
```

```python
path = "../data/"
f = xr.open_dataset(path + "postprc_map_month_am4.nc").load()
f.close()

lon_plot = f.lon.values
lat_plot = f.lat.values

ice = f.isel(season = 0).ice_mask.values
sice = f.isel(season = 1).ice_mask.values

cllcalipso = f.isel(season = 0).cllcalipso_val.values
scllcalipso = f.isel(season = 1).cllcalipso_val.values

cllcalipso_am4 = f.isel(season = 0).cllcalipso.values
scllcalipso_am4 = f.isel(season = 1).cllcalipso.values

cll_am4 = f.isel(season = 0).low_cld_amt.values
scll_am4 = f.isel(season = 1).low_cld_amt.values
```

```python
myfontsize = 12

fig, axs = plt.subplots(ncols = 3,
                        nrows = 2,
#                         sharex = True, sharey = True,
                        figsize=(21/2.54, (12)/2.54), dpi = 300)
      
pltpolar(lon_plot, lat_plot, cllcalipso, 100.*ice, axs[0,0], mycmap = cm.viridis, mylabel ='CALIPSO',
         myfontsize = myfontsize, mylevels = np.linspace(40, 90., num=11))

pltpolar(lon_plot, lat_plot, cllcalipso_am4, 100.*ice, axs[0,1], mycmap = cm.viridis, mylabel ='AM4 COSP',
         myfontsize = myfontsize, mylevels = np.linspace(40, 90., num=11))

pltpolar(lon_plot, lat_plot, cll_am4, 100.*ice, axs[0,2], mycmap = cm.viridis, mylabel ='AM4',
         myfontsize = myfontsize, mylevels = np.linspace(40, 90., num=11))

pltpolar(lon_plot, lat_plot, scllcalipso, 100.*sice, axs[1,0], mycmap = cm.viridis,
         myfontsize = myfontsize, mylevels = np.linspace(40, 90., num=11), hemisphere = 'sh')

pltpolar(lon_plot, lat_plot, scllcalipso_am4, 100.*sice, axs[1,1], mycmap = cm.viridis,
         myfontsize = myfontsize, mylevels = np.linspace(40, 90., num=11), hemisphere = 'sh')

oax= pltpolar(lon_plot, lat_plot, scll_am4, 100.*sice, axs[1,2], mycmap = cm.viridis,
         myfontsize = myfontsize, mylevels = np.linspace(40, 90., num=11), hemisphere = 'sh')

cbar = fig.colorbar(oax,  ax=axs, orientation='vertical', pad=0.03, shrink = 0.5, extend='both')
cbar.set_label(r"Low cloud fraction (%)", fontsize=myfontsize)

mylabel = np.array((['a','b','c'],['d','e','f']))

for i in range(2):
    for j in range (3):
        axs[i,j].text(0.05, 1.05, mylabel[i,j], transform=axs[i,j].transAxes,fontsize=myfontsize, fontweight='bold')
        
fig.savefig('Fig2_map', dpi=fig.dpi, bbox_inches='tight')
```
