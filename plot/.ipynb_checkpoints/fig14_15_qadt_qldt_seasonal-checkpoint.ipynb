```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import xarray as xr
import glob
from plot_utils import plt2D
```

```python
path = "../data/am4/"
year = '20100101-20141231'
exp = 'c96L33_am4p0_qadt_cosp_2010'
hemisphere = 'nh'

f = xr.open_mfdataset(path + hemisphere + '_' + year + "_average_transect_" + exp +"_all.nc",decode_cf = True).load()
f.close()

f = f.sel(distance = slice(-6.5, 6.5))

myfontsize = 12
```

```python
myfontsize = 12
fig, axs = plt.subplots(ncols = 3,
                        nrows = 3,
                        sharex = True, sharey = True,
                        figsize=(22/2.54, 12/2.54), dpi = 300)

plt.tight_layout()
lim = 4.*(10.**-5)
ncl = 10
mylevels1 = [35., 40., 45., 50., 70]

#--------------1st col-----------------------------------------------------------
ff0 = f.isel(season = 2)
x = ff0.distance.values
y = ff0.level.values
cld_amt = 100.*ff0.cld_amt.values

var = ff0.qadt_conv
plt2D(fig, axs[0,0], x, y, var, z1 = cld_amt, title = 'Spring \n convection', levels=np.linspace(-lim, lim, ncl + 1), 
      ylabel = 'Pres. (hPa)', cbar = False, cmap=cm.bwr, levels1 = mylevels1)

#
var = ff0.qadt_lsform
plt2D(fig, axs[1,0], x, y, var, z1 = cld_amt, title = 'large-scale cond.', levels=np.linspace(-lim, lim, ncl + 1), 
      ylabel = 'Pres. (hPa)', cbar = False, cmap=cm.bwr, levels1 = mylevels1)

#
var = ff0.qadt_super
plt2D(fig, axs[2,0], x, y, var, z1 = cld_amt, title = 'supersat. adjust.', levels=np.linspace(-lim, lim, ncl + 1), 
      ylabel = 'Pres. (hPa)', cbar = False, cmap=cm.bwr, levels1 = mylevels1)

#--------------2nd col-----------------------------------------------------------
ff0 = f.isel(season = 3)
x = ff0.distance.values
y = ff0.level.values
y1 = y
cld_amt = 100.*ff0.cld_amt.values

var = ff0.qadt_conv
plt2D(fig, axs[0,1], x, y, var, z1 = cld_amt, title = 'Autumn \n convection', levels=np.linspace(-lim, lim, ncl + 1),
     cbar = False, cmap=cm.bwr, levels1 = mylevels1)

#
var = ff0.qadt_lsform
plt2D(fig, axs[1,1], x, y, var, z1 = cld_amt, title = 'large-scale cond.', levels=np.linspace(-lim, lim, ncl + 1),
     cbar = False, cmap=cm.bwr, levels1 = mylevels1)

#
var = ff0.qadt_super
plt2D(fig, axs[2,1], x, y, var, z1 = cld_amt, title = 'supersat. adjust.', levels=np.linspace(-lim, lim, ncl + 1),
     cbar = False, cmap=cm.bwr, levels1 = mylevels1, xlabel = True)

#--------------3rd col-----------------------------------------------------------
ff0 = f.isel(season = 1)
x = ff0.distance.values
y = ff0.level.values
y1 = y
cld_amt = 100.*ff0.cld_amt.values

var = ff0.qadt_conv
plt2D(fig, axs[0,2], x, y, var, z1 = cld_amt, title = 'Summer \n convection', levels=np.linspace(-lim, lim, ncl + 1),
     cbar = False, cmap=cm.bwr, levels1 = mylevels1)

#
var = ff0.qadt_lsform
plt2D(fig, axs[1,2], x, y, var, z1 = cld_amt, title = 'large-scale cond.', levels=np.linspace(-lim, lim, ncl + 1),
     cbar = False, cmap=cm.bwr, levels1 = mylevels1)

#
var = ff0.qadt_super
oax = plt2D(fig, axs[2,2], x, y, var, z1 = cld_amt, title = 'supersat. adjust.', levels=np.linspace(-lim, lim, ncl + 1),
     cbar = False, cmap=cm.bwr, levels1 = mylevels1)

cbar = fig.colorbar(oax, ax = axs, orientation='vertical', 
                    shrink=0.5, pad = 0.05,ticks = [-lim, 0, lim])
cbar.set_label('Tendency of cloudiness ($s^{-1}$)', fontsize=myfontsize)

mylabel = np.array((['a','b','c'],['d','e','f'], ['h','i','j']))

for i in range(3):
    for j in range (3):
        axs[i,j].text(0.05, 0.85, mylabel[i,j], transform=axs[i,j].transAxes,fontsize=myfontsize, fontweight='bold')

        
fig.savefig('fig14_qadt_seasonal.png', dpi=fig.dpi, bbox_inches='tight')
```

```python
myfontsize = 12
fig, axs = plt.subplots(ncols = 3,
                        nrows = 3,
                        sharex = True, sharey = True,
                        figsize=(23/2.54, 12/2.54), dpi = 300)

plt.tight_layout()
lim = 10.*(10.**-9)
ncl = 10
mylevels1 = [1., 2., 3., 4.]

#--------------1st col-----------------------------------------------------------
ff0 = f.isel(season = 2)
x = ff0.distance.values
y = ff0.level.values
y1 = y
cld_amt = (10.**5)*(ff0.liq_wat.values+ff0.ice_wat.values)

var = ff0.qldt_conv
plt2D(fig, axs[0,0], x, y, var, z1 = cld_amt, title = 'Spring \n convection', levels=np.linspace(-lim, lim, ncl + 1), 
      ylabel = 'Pres. (hPa)', cbar = False, cmap=cm.bwr, levels1 = mylevels1)

#
var = ff0.qldt_cond
plt2D(fig, axs[1,0], x, y, var, z1 = cld_amt, title = 'large-scale cond.', levels=np.linspace(-lim, lim, ncl + 1), 
      ylabel = 'Pres. (hPa)', cbar = False, cmap=cm.bwr, levels1 = mylevels1)

#
var = ff0.qldt_liqadj
plt2D(fig, axs[2,0], x, y, var, z1 = cld_amt, title = 'supersat. adjust.', levels=np.linspace(-lim, lim, ncl + 1), 
      ylabel = 'Pres. (hPa)', cbar = False, cmap=cm.bwr, levels1 = mylevels1)


#--------------2nd col-----------------------------------------------------------
ff0 = f.isel(season = 3)
cld_amt = (10.**5)*(ff0.liq_wat.values+ff0.ice_wat.values)

var = ff0.qldt_conv
plt2D(fig, axs[0,1], x, y, var, z1 = cld_amt, title = 'Autumn \n convection', levels=np.linspace(-lim, lim, ncl + 1),
      cbar = False, cmap=cm.bwr, levels1 = mylevels1)

#
var = ff0.qldt_cond
plt2D(fig, axs[1,1], x, y, var, z1 = cld_amt, title = 'large-scale cond.', levels=np.linspace(-lim, lim, ncl + 1),
      cbar = False, cmap=cm.bwr, levels1 = mylevels1)

#
var = ff0.qldt_liqadj
plt2D(fig, axs[2,1], x, y, var, z1 = cld_amt, title = 'supersat. adjust.', levels=np.linspace(-lim, lim, ncl + 1),
      cbar = False, cmap=cm.bwr, levels1 = mylevels1, xlabel = True )

#--------------3rd col-----------------------------------------------------------
ff0 = f.isel(season = 1)
cld_amt = (10.**5)*(ff0.liq_wat.values+ff0.ice_wat.values)

var = ff0.qldt_conv
plt2D(fig, axs[0,2], x, y, var, z1 = cld_amt, title = 'Summer \n convection', levels=np.linspace(-lim, lim, ncl + 1),
      cbar = False, cmap=cm.bwr, levels1 = mylevels1)

#
var = ff0.qldt_cond
plt2D(fig, axs[1,2], x, y, var, z1 = cld_amt, title = 'large-scale cond.', levels=np.linspace(-lim, lim, ncl + 1),
      cbar = False, cmap=cm.bwr, levels1 = mylevels1)

#
var = ff0.qldt_liqadj
oax = plt2D(fig, axs[2,2], x, y, var, z1 = cld_amt, title = 'supersat. adjust.', levels=np.linspace(-lim, lim, ncl + 1),
           cbar = False, cmap=cm.bwr, levels1 = mylevels1)


cbar = fig.colorbar(oax, ax = axs, orientation='vertical', 
                    shrink=0.5, pad = 0.05,ticks = [-lim, 0, lim])
cbar.set_label('Tendency of cloud liquid water ($kgkg^{-1}s^{-1}$)', fontsize=myfontsize)

mylabel = np.array((['a','b','c'],['d','e','f'], ['h','i','j']))

for i in range(3):
    for j in range (3):
        axs[i,j].text(0.05, 0.85, mylabel[i,j], transform=axs[i,j].transAxes,fontsize=myfontsize, fontweight='bold')


fig.savefig('Fig15_qldt_seasonal.png', dpi=fig.dpi, bbox_inches='tight')
```
