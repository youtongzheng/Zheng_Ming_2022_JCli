```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import xarray as xr
import glob
```

```python
path = "../data/am4/"
year = '20100101-20141231'
exp = 'c96L33_am4p0_qadt_cosp_2010'


f = xr.open_mfdataset(path + 'nh_' + year + "_average_transect_" + exp +"_all.nc",decode_cf = True).load()
f.close()
```

```python
pres = f.level.expand_dims(distance = f['distance'].size).transpose().expand_dims(season = f['season'].size)
f['lts'] = (f.temp*(1000./pres)**(0.286)).isel(level = 3) - (f.temp*(1000./pres)**(0.286)).isel(level = 15) 
f['ts_ta'] = f.t_surf- f.temp.isel(level = 15)
```

```python
myfontsize = 12
fig, axs = plt.subplots(ncols = 2,
                        nrows = 2,
                        sharex = True, sharey = False,
                        figsize=(14/2.54, 9/2.54), dpi = 300)

plt.tight_layout()

#fig1
x = f.distance.values
axs[0,0].plot(x, f.t_surf.isel(season = 0).values, label = 'Winter')
axs[0,0].plot(x, f.t_surf.isel(season = 1).values, label = 'Summer')
axs[0,0].plot(x, f.t_surf.isel(season = 2).values, label = 'Spring')
axs[0,0].plot(x, f.t_surf.isel(season = 3).values, label = 'Autumn')

axs[0,0].set_xlim([-6, 6])

axs[0,0].set_ylabel('$\mathrm{T_{sfc}}$ (K)', fontsize=myfontsize) 
# axs[0,0].set_xlabel('Distance from ice edge \n ($\degree$ latitude)', fontsize=myfontsize) 
axs[0,0].axvline(x=0., color='black', lw = 0.5, linestyle='--')
axs[0,0].axhline(y=273.15, color='grey', lw = 0.5, linestyle='--')
axs[0,0].legend(loc = 'best',fontsize=0.7*myfontsize)

#fig2
axs[0,1].plot(x, f.ts_ta.isel(season = 0).values, label = 'Winter')
axs[0,1].plot(x, f.ts_ta.isel(season = 1).values, label = 'Summer')
axs[0,1].plot(x, f.ts_ta.isel(season = 2).values, label = 'Spring')
axs[0,1].plot(x, f.ts_ta.isel(season = 3).values, label = 'Autumn')

axs[0,1].set_ylabel('$\mathrm{T_{sfc} - T_a}$ (K)', fontsize=myfontsize) 
# axs[0,1].set_xlabel('Distance from ice edge \n ($\degree$ latitude)', fontsize=myfontsize) 
axs[0,1].axvline(x=0., color='black', lw = 0.5, linestyle='--')

#fig3
axs[1,0].plot(x, f.lts.isel(season = 0).values, label = 'Winter')
axs[1,0].plot(x, f.lts.isel(season = 1).values, label = 'Summer')
axs[1,0].plot(x, f.lts.isel(season = 2).values, label = 'Spring')
axs[1,0].plot(x, f.lts.isel(season = 3).values, label = 'Autumn')

axs[1,0].set_ylabel('LTS (K)', fontsize=myfontsize) 
axs[1,0].set_xlabel('Distance from ice edge \n ($\degree$ latitude)', fontsize=myfontsize) 
axs[1,0].axvline(x=0., color='black', lw = 0.5, linestyle='--')

#fig4
axs[1,1].plot(x, f.evap.isel(season = 0).values, label = 'Winter')
axs[1,1].plot(x, f.evap.isel(season = 1).values, label = 'Summer')
axs[1,1].plot(x, f.evap.isel(season = 2).values, label = 'Spring')
axs[1,1].plot(x, f.evap.isel(season = 3).values, label = 'Autumn')

axs[1,1].set_ylabel('sfc. evap. $\mathrm{(kgm^{-2}s^{-1})}$', fontsize=myfontsize) 
axs[1,1].set_xlabel('Distance from ice edge \n ($\degree$ latitude)', fontsize=myfontsize) 
axs[1,1].axvline(x=0., color='black', lw = 0.5, linestyle='--')

mylabel = np.array((['a','b'],['c','d']))

for i in range(2):
    for j in range (2):
        axs[i,j].text(0.05, 0.85, mylabel[i,j], transform=axs[i,j].transAxes,fontsize=myfontsize, fontweight='bold')
        
fig.savefig('Fig16_ts_lts_lhf_ts-ta.png', dpi=fig.dpi, bbox_inches='tight')
```
