```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import xarray as xr
import glob
from plot_utils import plt2D
```

```python
path = "../data/am4/"
year = '20100101-20141231'
exp = 'c96L33_am4p0_qadt_cosp_2010'
hemisphere = 'nh'

f = xr.open_mfdataset(path + hemisphere + '_' + year + "_average_transect_" + exp +"_all.nc",decode_cf = True).load()
f.close()

f = f.sel(distance = slice(-6.5, 6.5))

myfontsize = 12
```

```python
fig, axs = plt.subplots(ncols = 3,
                        nrows = 2,
                        sharex = True, sharey = True,
                        figsize=(23/2.54, 8/2.54), dpi = 300)

plt.tight_layout()

ff0 = f.isel(season = 0)
lim = 4.*(10.**-5)
ncl = 10

x = ff0.distance.values
y = ff0.level.values
cld_amt = 100.*ff0.cld_amt.values

#1
var = ff0.qadt_conv
plt2D(fig, axs[0,0], x, y, var, z1 = cld_amt, title = '(a) convection', levels=np.linspace(-lim, lim, ncl + 1), 
      ylabel = 'Pres. (hPa)', cbar = False, cmap=cm.bwr)

#2
var = ff0.qadt_lsform 
plt2D(fig, axs[0,1], x, y, var, z1 = cld_amt, title = '(b) large-scale cond.', levels=np.linspace(-lim, lim, ncl + 1),
      cbar = False, cmap=cm.bwr)

#3
var = ff0.qadt_super
plt2D(fig, axs[0,2], x, y, var, z1 = cld_amt, title = '(c) supersat. adjust.', levels=np.linspace(-lim, lim, ncl + 1),
      cbar = False, cmap=cm.bwr)

#4
var = ff0.qadt_eros
plt2D(fig, axs[1,0], x, y, var, z1 = cld_amt, title = '(d) erosion', levels=np.linspace(-lim, lim, ncl + 1), 
      ylabel = 'Pres. (hPa)',cbar = False, cmap=cm.bwr)

#5
var = ff0.qadt_vdif
plt2D(fig, axs[1,1], x, y, var, z1 = cld_amt, title = '(e) vert. diffusion', levels=np.linspace(-lim, lim, ncl + 1), 
      xlabel = True, cbar = False, cmap=cm.bwr)

#6
var = ff0.qadt_dyn
oax = plt2D(fig, axs[1,2], x, y, var, z1 = cld_amt, title = '(f) advection', levels=np.linspace(-lim, lim, ncl + 1),
     cbar = False, cmap=cm.bwr)

cbar = fig.colorbar(oax, ax = axs, orientation='vertical', 
                    shrink=0.6, pad = 0.05,ticks = [-lim, 0, lim])
cbar.set_label('Tendency of cloudiness ($s^{-1}$)', fontsize=myfontsize)

# mylabel = np.array((['a','b','c'],['d','e','f']))

# mylabel = np.array((['a convection','b large-scale cond.','c boundary-layer cond.'],
#                     ['d erosion','e vert. diffusion','f advection']))

# for i in range(2):
#     for j in range (3):
#         axs[i,j].text(0.05, 0.85, mylabel[i,j], transform=axs[i,j].transAxes,
#                       fontsize=myfontsize, fontweight='bold')

fig.savefig('Fig8_qadt.png', dpi=fig.dpi, bbox_inches='tight')
```
